```python
print("Operaciones disponibles: suma, resta, multiplicacion, division, potencia")
print("Puedes ingresar números (10) o colecciones ([1, 5], {10, 20}, (3, 3))")

operacion_str = input("Seleccione la operación: ").lower()

num1_str = input("Ingrese el primer valor: ")
num2_str = input("Ingrese el segundo valor: ")

try:
    num1_obj = eval(num1_str)
    num2_obj = eval(num2_str)

    if isinstance(num1_obj, (list, set, tuple)):
        num1 = sum(num1_obj)
    else:
        num1 = int(num1_obj) # Si no, lo tratamos como un número normal

    if isinstance(num2_obj, (list, set, tuple)):
        num2 = sum(num2_obj)
    else:
        num2 = int(num2_obj)

except (SyntaxError, NameError, TypeError, ValueError):
    print("ERROR: Entrada inválida. Asegúrate de ingresar números o colecciones válidas.")
    exit()

resultado = 0 
simbolo = "" 

if operacion_str == 'suma':
    resultado = num1 + num2
    simbolo = "+"
    
elif operacion_str == 'resta':
    resultado = num1 - num2
    simbolo = "-"
    
elif operacion_str == 'multiplicacion':
    resultado = num1 * num2
    simbolo = "*"
    
elif operacion_str == 'division':
    if num2 != 0:
        resultado = num1 / num2
        simbolo = "/"
    else:
        resultado = "ERROR: División por cero"
        simbolo = "/"
        
elif operacion_str == 'potencia':
    resultado = num1 ** num2
    simbolo = "**"
    
else:
    resultado = "ERROR: Operación no reconocida"
    simbolo = "?"


print(f"Valor 1 Ingresado: {num1_str} ")
print(f"Valor 2 Ingresado: {num2_str} ")
print(f"Operación Solicitada: {operacion_str.capitalize()}")

if isinstance(resultado, str):
    print(f"Resultado: {resultado}")
else:
    print(f"El cálculo es: {num1} {simbolo} {num2} = {resultado}")
```
